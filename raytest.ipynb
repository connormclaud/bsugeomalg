```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
```

```python
vertices = np.array([[0.01362163, 0.65575358],
       [0.76353171, 0.64278502],
       [0.94599151, 0.21497581],
       [0.65417826, 0.25345748],
       [0.28248603, 0.22315736],
       [0.52071828, 0.38373991],
       [0.46265165, 0.58939919]])
```

```python
fig2 = plt.figure()
ax2 = fig2.add_subplot(111, aspect='equal')

poly = patches.Polygon(vertices, True,
        fill=False      # remove background
    )
ax2.add_patch(
    poly
)
x = np.random.rand(1)
y = np.random.rand(1)
point = (*x, *y)
ax2.scatter(x, y)
plt.show()
```

```python
def bound_test(point, vertices):
    minx, miny = tuple(vertices.min(axis=0))
    maxx, maxy = tuple(vertices.max(axis=0))
    if minx <= point[0] <= maxx and miny <= point[1] <= maxy:
        return True
    else:
        return False
```

```python
def ray_test(point, vertices):
    if not bound_test(point, vertices):
        return False
    ray_start = (point[0], 1)
    ray_end = (point[1], point[1])
    ax2.plot(ray_start, ray_end, 'g-')
    edges = zip(vertices, vertices[1:] + vertices[:1])
    print(vertices)
    for A, B in edges:
        print(A, B)
        ax2.plot((A[0], B[0]), (A[1], B[1]), 'r--')
```
